*note: only work if page number is x / y as the last thing on the slides, if not will not read the "x" properly and have to change code a bit

```python
#creating a pdf file object
#------------------------------------------------------------------------------------
mypdf = open("/content/PBE once again.pdf", 'rb') #ENTER PDF FILE PATH TO BE READ HERE
#------------------------------------------------------------------------------------
# creating a pdf reader object
reader = PyPDF2.PdfReader(mypdf)

parts=[]
numbarray = []
def visitor_body(text, cm, tm, fontDict, fontSize):
    y = tm[5]
    if y < 50:
        parts.append(text)



#make list of all pages
for i in range(len(reader.pages)):
  page = reader.pages[i]
  page.extract_text(visitor_text=visitor_body)
  text_body = "".join(parts)
  pagenumbs = text_body.split()
  pagenumber = pagenumbs[-3] #note this is what to change if slide number is different format
  #print(pagenumber)
  numbarray.append(pagenumber)


writer = PdfWriter() #defines a pdf to write in

#adds greatest nonrepetitive page to writer
numbskip=99999999999
for i in range(len(numbarray)):
  if numbskip == numbarray[i]:
    #dont go through, alredy got the greatest
    continue
  else:
    for j in range(len(numbarray)):
      if numbarray[i] == numbarray[len(numbarray)-j-1]:
        numbskip = numbarray[len(numbarray)-j-1]
        writer.add_page(reader.pages[len(numbarray)-j-1])
        break
      else:
        continue
    #go through for loop again, check for next lowest number


#NAME THE DOCUMENT HERE
#------------------------------------------------------------------------------------
with open("PBE once again.pdf", "wb") as fp:
#------------------------------------------------------------------------------------
    writer.write(fp)

# closing the pdf file object
mypdf.close()
```
